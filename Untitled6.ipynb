```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train, X_test = scaler.fit_transform(X_train), scaler.fit_transform(X_test)
```

```python
regr = DecisionTreeRegressor(max_depth=3)
model = regr.fit(X_train, y_train)
regr_predict = regr.predict(X_test)
regr_predict.shape
```

```python
text_representation = tree.export_text(regr)
feature_names = fetch_california_housing().feature_names
fig = plt.figure(figsize=(30,20))
graph = tree.plot_tree(regr, feature_names=feature_names, filled=True)
plt.show()
# https://practicaldatascience.co.uk/machine-learning/how-to-use-model-selection-and-hyperparameter-tuning
```

```python
model.get_params()
```

```python
param_grid = dict(
ccp_alpha=[0,1],
max_depth=[3,50,90,150,200],
min_samples_leaf=[20,40,60,100],
min_samples_split=[2,3,40,60,100])
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_root_mean_squared_error')
```

```python
best_model = grid_search.fit(X_train, y_train)
```

```python
best_model.best_params_
```

```python
model_test = DecisionTreeRegressor(max_depth=50, min_samples_leaf=20, min_samples_split=60)
```

```python
model_test.fit(X_train, y_train)
y_pred = model_test.predict(X_test)
```

```python
X_train
```

```python
test = pd.DataFrame({'Predicted value':y_pred, 'Actual value':y_test})
fig= plt.figure(figsize=(16,8))
test = test.reset_index()
test = test.drop(['index'], axis=1)
plt.plot(test[:100])
plt.legend(['Predicted value','Actual value'])
plt.xlabel("index")
plt.ylabel("y_vals")
plt.show()
```
